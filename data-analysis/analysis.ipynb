```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("../data/realdonaldtrump.csv")
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
from ttp import ttp
p = ttp.Parser()
result = p.parse(df.loc[0, "content"])
print(result.reply)
print(result.users)
print(result.tags)
print(result.urls)
```

```python
most_frequent_tweets = df['content'].value_counts().rename_axis('tweet_text').reset_index(name='count')[0:10]
most_frequent_tweets.insert(0, 'id', range(0, 0 + len(most_frequent_tweets)))
most_frequent_tweets.head()
```

```python
most_frequent_tweets.to_csv('../data/most_frequent_tweets.csv', index=True, header=True)
```

```python
df.insert(8, 'has_exclaimation', df['content'].str.count("!"))
df.head()
```

```python
df['has_exclaimation'].describe()
```

```python
cut_labels = ["one","two","three","4-5","6-8","9-10", ">10"]
cut_bins = [0,1,2,3,5,8,10,17]

df['exclaimation_bin'] = pd.cut(df['has_exclaimation'], bins=cut_bins, labels=cut_labels).values

df['exclaimation_bin'].describe()
df.to_csv('../data/has_exclaimation.csv', index=True, header=True)
```

```python
df['hour'] = df.groupby( [df.date.dt.hour]).grouper.group_info[0]
df.head()
```

```python
df.to_csv('../data/tweets_by_hour.csv', index=True, header=True)
```
